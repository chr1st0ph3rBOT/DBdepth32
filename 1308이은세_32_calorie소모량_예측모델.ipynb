# 데이터 시각화를 통한 각 Feature간 상관관계 분석(이해)하기
* 아래의 시각화 결과를 어떻게 해석하고 모델링에 적용 또는 결과 분석에 활용할 것인가?

```python
# 데이터 시각화에 필요한 matplotlib, seaborn 라이브러리를 가져오는 코드를 작성해보자. 각각 별명(칭)은 plt, sns이다.
# 유의할 점: matplotlib을 가져올 때에는 특히, pyplot이라는 하위 모듈만을 가져와야한다.


features = ['Gender', 'Age', 'Height', 'Weight', 'Duration', 'Heart_Rate', 'Body_Temp']

plt.figure(figsize=(18, 12))

# ★ for ~ in enumerate()의 구조에서 enumerate()가 하는 일은? range와 다르게 어떤 차이가 있을까?
for i, feature in enumerate(features):
    plt.subplot(3, 3, i+1)  # plt.subplot의 인덱스는 1부터 시작함.
    sns.scatterplot(data=merged_df, x=feature, y='Calories')
    plt.title(f'{feature} vs Calories')

#plt.tight_layout() # 해당 명령문을 주석처리했을 경우와 실행했을 경우의 차이를 확인하자!
plt.show()
```

**#히트맵(HeatMap)의 의미 해석하기**
- 히트맵이란? 데이터의 패턴, 상관 관계 및 분포를 시각적으로 표현하기 위해 주로 사용되는 강력한 시각화 도구 중 하나로, 주로 2차원 데이터를 다루며, 데이터를 행과 열로 구성된 행렬 형태로 나타냄.
[출처] https://wikidocs.net/216468
- 양의 상관계수 vs 음의 상관계수
- 색상은 상관계수(Pearson correlation coefficient)를 시각적으로 표현함
**   1.00에 가까울수록 강한 양의 상관관계 (ex: 운동 시간과 칼로리)
**   -1.00에 가까울수록 강한 음의 상관관계
**   0에 가까울수록 거의 무관함을 의미함
- Duration, Heart_Rate, Body_Temp는 Calories와 강한 양의 상관관계를 보임.
- Gender, Height, Age 등은 칼로리 소모량과 상대적으로 약한 관계를 보임.

* https://seaborn.pydata.org/generated/seaborn.heatmap.html

```python
# 상관계수 행렬 계산 (User_ID는 제거한 후, 상관계수 계산 corr())
correlation_matrix = merged_df.drop(columns=['User_ID']).corr()

# 히트맵 시각화
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.show()
```
